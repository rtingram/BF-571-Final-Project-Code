```python
import pandas as pd
import re
from bokeh.plotting import figure, output_file, show, ColumnDataSource
from bokeh.models import HoverTool, Range1d
from bokeh.io import output_notebook
output_notebook()
```

```python
## Load the patient data
# Viral Load
p1_vload = pd.read_csv('viralLoad_p1.tsv', sep='\t')
p2_vload = pd.read_csv('viralLoad_p2.tsv', sep='\t')
p3_vload = pd.read_csv('viralLoad_p3.tsv', sep='\t')
p5_vload = pd.read_csv('viralLoad_p5.tsv', sep='\t')
p6_vload = pd.read_csv('viralLoad_p6.tsv', sep='\t')
p7_vload = pd.read_csv('viralLoad_p7.tsv', sep='\t')
p8_vload = pd.read_csv('viralLoad_p8.tsv', sep='\t')
p9_vload = pd.read_csv('viralLoad_p9.tsv', sep='\t')
p10_vload = pd.read_csv('viralLoad_p10.tsv', sep='\t')
p11_vload = pd.read_csv('viralLoad_p11.tsv', sep='\t')

# Cell Counts
p1_cd = pd.read_csv('cellCount_p1.tsv', sep='\t')
p2_cd = pd.read_csv('cellCount_p2.tsv', sep='\t')
p3_cd = pd.read_csv('cellCount_p3.tsv', sep='\t')
p5_cd = pd.read_csv('cellCount_p5.tsv', sep='\t')
p6_cd = pd.read_csv('cellCount_p6.tsv', sep='\t')
p7_cd = pd.read_csv('cellCount_p7.tsv', sep='\t')
p8_cd = pd.read_csv('cellCount_p8.tsv', sep='\t')
p9_cd = pd.read_csv('cellCount_p9.tsv', sep='\t')
p10_cd = pd.read_csv('cellCount_p10.tsv', sep='\t')
p11_cd = pd.read_csv('cellCount_p11.tsv', sep='\t')
```

```python
# Days from Infection
xs = [p1_cd['Days from infection'], p2_cd['Days from infection'], p3_cd['Days from infection'], 
      p5_cd['Days from infection'], p6_cd['Days from infection'], p7_cd['Days from infection'],
      p8_cd['Days from infection'], p9_cd['Days from infection'], p10_cd['Days from infection'],
      p11_cd['Days from infection']]

# CD4+ Cell Count
ys = [p1_cd['CD4+ count [cells/ul]'], p2_cd['CD4+ count [cells/ul]'], p3_cd['CD4+ count [cells/ul]'], 
      p5_cd['CD4+ count [cells/ul]'], p6_cd['CD4+ count [cells/ul]'], p7_cd['CD4+ count [cells/ul]'],
      p8_cd['CD4+ count [cells/ul]'], p9_cd['CD4+ count [cells/ul]'], p10_cd['CD4+ count [cells/ul]'],
      p11_cd['CD4+ count [cells/ul]']]


# Plotting Function
def make_bokeh_plot():
    # Hover tool for interactive plots
    hover = HoverTool(
            tooltips=[
                ("index", "$index"),
                ("(x,y)", "($x, $y)"),
                ("desc", "@desc"),
            ]
        )

    # Creates log scale
    p = figure(tools=[hover], x_axis_type="log", y_axis_type="log")
    
    # Axis labes
    p.xaxis.axis_label = 'Days from infection'
    p.yaxis.axis_label = 'CD4+ count [cells/ul]'

    # Axis ranges
    p.x_range = Range1d(50, 10000)
    p.y_range = Range1d(125, 1550)

    # Axis formatting
    p.yaxis.major_label_orientation = "vertical"
    
    # Defined lists for 'for' loop
    color_list = ['red', 'green','blue','grey','gold','violet',
                  'black','brown','lightblue','lime']
    leg_list = ['P1', 'P2', 'P3','P5','P6','P7','P8','P9','P10','P11']
    
    # Loop that generates the legend
    for (colr, leg, x, y) in zip(color_list, leg_list, xs, ys):
        my_plot = p.multi_line(x, y, color= colr,legend=leg)

    #Main call that generates the plot
    p.multi_line(xs, ys, color=['red', 'green','blue','grey','gold','violet',
                                'black','brown','lightblue','lime'])

    #Show plot
    show(p)
```

```python
make_bokeh_plot()
```
